### 45. Shuffle the DataFrame using `sample()` with the frac parameter set to 1

Notice how the index numbers get moved around. The [`sample()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sample.html) function is a great way to get random samples from your DataFrame. It's also another great way to shuffle the rows by setting `frac=1`.
